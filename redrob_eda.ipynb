# EDA

````python
import json
import pandas as pd
from tqdm import tqdm

data = []

with open(CANDIDATE_DATA_FILE_PATH, "r") as f:

    for line in tqdm(f):

        candidate = json.loads(line)

        data.append(candidate)
````

````python
print("Total candidates:", len(data))
````

````python
data[0]
````

EDA on `data[i]['profile']['headline']`

````python
headline = set()

for i in range(len(data)):
    # 1. Grab the raw headline string from your data dictionary
    raw_text = data[i]['profile']['headline']

    # 2. Check if the text contains any numbers
    if any(char.isdigit() for char in raw_text):
        # Has a number: split at pipe, take the first part, and trim spaces
        cleaned = raw_text.split("|")[0].strip()
    else:
        # No number: replace the pipe with a comma
        cleaned = raw_text.replace(" |", ",").strip()

    # 3. Add the cleaned result to your set
    headline.add(cleaned)
````

````python
print(len(headline))
````

````python
for h in headline:
  print(h)
````

EDA on `data[i]['profile']['summary']`

````python
summary = set()

for i in range(len(data)):
    # 1. Grab the raw summary string from your data dictionary
    summary.add(data[i]['profile']['summary'])
````

````python
print(len(summary))
````

````python
summary = set()
seen_summaries = {}  # Map: {summary_text: candidate_id}
copy = 0

for i in range(len(data)):
    raw_text = data[i]['profile']['summary']
    current_id = data[i]['candidate_id']  # Get current candidate_id

    # Check if this summary text was already used by someone else
    if raw_text in seen_summaries:
        # print("--- Duplicate Found! ---")
        copy += 1

        # Retrieve the original candidate_id from our dictionary
        original_id = seen_summaries[raw_text]

        # print(f"Original Candidate ID: {original_id}")
        # print(f"Duplicate Candidate ID: {current_id}\n")
    else:
        # Save the summary text and pair it with the original candidate's ID
        seen_summaries[raw_text] = current_id

    summary.add(raw_text)

    # if copy > 5:
    #     break
````
